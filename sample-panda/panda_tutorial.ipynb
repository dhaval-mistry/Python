```python
#series 

s = pd.Series([1,2,3,4,5])
print(s)

#access by index
print(s[2])

#label
print(s[4])
```

```python
s = pd.Series([23,24,25,26], index=["aaa","bbb","ccc","ddd"])

print(s)

#access thru label 
print(s["aaa"])
print(s["ddd"])

#get index list 
print(s.index)
```

```python
s = pd.Series([1,2,3,4,5])

s[0] = 100
print(s)
#remove and assign to new Series  
changed_s = s.drop(0)

print(s)
print(changed_s)
```

```python
#DataFrame

info = {
    "Name" : ["Rahul","Harshita","Shradha"],
    "Age" : [23, 24, 25],
    "CGPA" : [9.5, 7.5, 8.5]
}

df = pd.DataFrame(info)

print(df)
#row index
print(df.index)
#column index
print(df.columns)
```
